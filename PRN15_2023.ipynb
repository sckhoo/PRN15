```python
states = list(prn15.state.unique())
parties = list(prn15.party.unique())
ethnicities = list(prn15.ethnicity.unique())
sexes = list(prn15.sex.unique())

acronym = list(prn15.acronym.unique())


print(states)
print(parties)
print(ethnicities)
print(sexes)
```

```python
for state in states:
  for party in parties:
    calculated_votes = prn15[(prn15['state'] == state) & (prn15['party'] == party)]['votes'].sum()
    print(f"{state} {party} {calculated_votes}")
```

```python
prn15.loc[prn15.dun == 'N.31 Subang Jaya']
```

```python
for state in states:
  for party in parties:
    min_age = prn15[(prn15['state'] == state) & (prn15['party'] == party)]['age'].min()
    med_age = prn15[(prn15['state'] == state) & (prn15['party'] == party)]['age'].median()
    max_age = prn15[(prn15['state'] == state) & (prn15['party'] == party)]['age'].max()
    print(f"{state} {party} {min_age} {med_age} {max_age}")
```

```python
for state in states:
  for ethnicity in ethnicities:
    min_age = prn15[(prn15['state'] == state) & (prn15['ethnicity'] == ethnicity)]['age'].min()
    med_age = prn15[(prn15['state'] == state) & (prn15['ethnicity'] == ethnicity)]['age'].median()
    max_age = prn15[(prn15['state'] == state) & (prn15['ethnicity'] == ethnicity)]['age'].max()
    print(f"{state} {ethnicity} {min_age} {med_age} {max_age}")
```

```python
female, male = [], []
for state in states:
  for sex in sexes:
    count =  len(prn15[(prn15['state'] == state) & (prn15['sex'] == sex)]['sex'])
    if sex == 'female':
      female.append(count)
    else:
      male.append(count)
    print(f"{state} {sex} {count}")
print(female)
print(male)
```

```python
df = pd.DataFrame({'female':female, 'male':male}, index=states)
df.plot.bar()
```

```python
ages = [[],[],[],[],[],[]]
#for n, state in enumerate(states):
#  ages[n] = []
for n, state in enumerate(states):
  age = prn15[(prn15['state'] == state)]['age']
  ages[n] = age.to_list()
  print(len(ages[n]))
#print(ages)
```

```python
df = pd.DataFrame({'data': [ages[0], ages[1], ages[2], ages[3], ages[4], ages[5]]}) #, 'label': states})

plt.boxplot(df['data'], labels=states, vert=False)
plt.xlabel('Candidate Age')

plt.show()
```

```python
state_gender = {}
for state in states:
  female_ratio = len(prn15[(prn15['state'] == state) & (prn15['sex'] == 'female')]['sex']) / len(prn15[(prn15['state'] == state)]['sex'])
  state_gender[state] = (female_ratio * 100)
print(state_gender)
```

```python
x_axis = list(state_gender.keys())
y_axis = list(state_gender.values())

plt.bar(x_axis, y_axis)
plt.xticks(x_axis, rotation=45)
plt.xlabel("State")
plt.ylabel("Female Percentage")
plt.show()
```

```python

# Calculate the total percentage of votes for each race in each state
race_data = prn15.groupby(["state", "ethnicity"])['name'].size()
race_data
```

```python
index = ['state', 'ethnicity', 'count']
race_data_df = pd.DataFrame(columns=['state', 'ethnicity', 'count'])
print(race_data_df)

for state in states:
  for ethnicity in ethnicities:
    print(f"{state} {ethnicity} {race_data.get(state).get(ethnicity)}")
    temp_df = pd.Series({'state':state,'ethnicity':ethnicity,'count':race_data.get(state).get(ethnicity)})
    #print(temp_df)
    race_data_df = pd.concat([race_data_df, temp_df.to_frame().T], ignore_index=True)
race_data_df = race_data_df.fillna(0)
race_data_df
```

```python
race_data_df_pivot = pd.pivot_table(race_data_df, index='state', values='count', columns=['ethnicity'], sort=False)
race_data_df_pivot
```

```python
race_data_df_pivot.plot.bar(stacked=True) #plt.xticks(x_axis, rotation=45)
```
